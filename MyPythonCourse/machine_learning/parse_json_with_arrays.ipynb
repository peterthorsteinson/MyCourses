```python
import json

input_file = open ('stores-small.json')
json_array = json.load(input_file)
store_list = []

for item in json_array:
    store_details = {"name":None, "city":None}
    store_details['name'] = item['name']
    store_details['city'] = item['city']
    store_details['location'] = item['location']
    store_details['services'] = item['services']
    store_list.append(store_details)
print(store_list)
print()
print(store_details['services'])
```

```python
location = str(store_details['location'])
print(location)
data = json.loads(location)
for element in data['lat']:
    print(element)
```

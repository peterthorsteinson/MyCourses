# Logistic Map Chaos

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def valsByTime (k, x, nSteps) :
    tims = range(nSteps)
    vals = []
    for i in tims :
        xp = nextX(k,x)
        vals.append(xp)
        x = xp
    return tims, vals

def nextX (k,x) :
    return x * (1.0-x) * k
```

```python
def plotLogisticMap(k):
    plt.suptitle("k is %f" % k, size=16)
    tims, vals = valsByTime(k, 0.35, 50)
    plt.plot(tims, vals, 'r')
    tims, vals = valsByTime(k, 0.34, 50)
    plt.plot(tims, vals, 'b')
    plt.show()
```

```python
plotLogisticMap(k=1.0)
```

```python
plotLogisticMap(k=2.7)
```

```python
plotLogisticMap(k=2.9)
```

```python
plotLogisticMap(k=3.5)
```

```python
plotLogisticMap(k=3.7)
```

## Logisctic Map Bifurcation Animation Diagram

**NOTE: In Jupyter notebook, tkinter displays GUI output in seperate window, not inline within the notebook itself**

```python
# Logisctic Map Bifurcation Diagram

from tkinter import Tk, Canvas, Frame
    
canWidth=500
canHeight=500

def setupWindow () :
    global win, canvas
    win = Tk()
    canvas = Canvas(win, height=canHeight, width=canWidth)
    f = Frame (win)    
    canvas.pack()
    f.pack()

def startApp () :
    global win, canvas
    k1  = float(2.0)   # starting value of K
    k2  = float(4.0)   # ending   value of K
    x = .2                     # is somewhat arbitrary
    vrng = range(200)          # We'll do 200 horz steps
    for t in range(canWidth) :
        win.update()
        k = k1 + (k2-k1)*t/canWidth
        #print("K = %.04f" % k)
        for i in vrng :
            p = x*canHeight
            canvas.create_line(t,p,t,p+1)  # just makes a pixel dot
            x = x * (1-x) * k              # next x value
            if x <=0 or x >= 1.0 :
                #print( "overflow at k", k)
                return
    
def main () :
    setupWindow()       # Create Canvas with Frame
    startApp()          # Start up the display  
    win.mainloop()      # Just wait for user to close graph
    
main()
```

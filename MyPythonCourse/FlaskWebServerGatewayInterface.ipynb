# Flask Web Server Gateway Interface (WSGI)

See: https://enterprise-docs.anaconda.com/en/latest/data-science-workflows/deployments/deploy-flask.html
See: https://www.fullstackpython.com/wsgi-servers.html

* Run the following cell and then open your browser to the URL: http://localhost:5000
* You should see the following HTML rendered in your browser:

```
<!DOCTYPE html>
<html>
<body>
<svg width="100" height="100">
  <circle cx="50" cy="50" r="40" stroke="blue" stroke-width="4" fill="yellow"/>
</svg>
</body>
</html>
```

````python
from flask import Flask, Blueprint

hello = Blueprint('hello', __name__)

@hello.route('/')
def hello_world():
    html = """
<!DOCTYPE html>
<html>
<body>
<svg width="100" height="100">
  <circle cx="50" cy="50" r="40" stroke="blue" stroke-width="4" fill="yellow"/>
</svg>
</body>
</html>
"""
    return html

if __name__ == '__main__':
    app = Flask(__name__)
    app.register_blueprint(hello, url_prefix='/')

    app.run()
````

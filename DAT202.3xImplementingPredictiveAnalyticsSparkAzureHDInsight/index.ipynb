### The following describes how to do it locally in a Docker container...

DOS Shell:
```
> docker run -t -p 8888:8888 jupyter/pyspark-notebook
```
        Displays 3 URLs -> view 3rd link in local browser

If you need a file share folder on host computer:
```
> docker run -t -p 8888:8888 -v C:\docker_share:/home/share jupyter/pyspark-notebook
```

If it prompts "Password or token" and the token in the URL does not work
then ?????

DOS Shell (new one):
```
> docker container list
        Copy container ID to clipboard
> docker exec -it container-id bash
        Paste in correct container ID in above command
```

Run the following bash command in the ubuntu container:

```
$ cd work
$ wget http://aka.ms//edx-dat202.3x-labs
$ unzip edx-dat202.3x-labs
$ pip install sparksql-magic
```

Open each notebook, such as -> work/Lab01/Python Data Exploration.ipynb

**Note** Original labs were in Python 2.x kernel, but you can run 3.X kernel if you update code, e.g. print -> print()

If you cange the Jupyter kernel from Python2 to Python3, make the change syntax changes: print ... -> print(...)

Add the following to the top cell in each notebook:

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local").getOrCreate()
```

Add the following to the top cell in each notebook that has any cell that contains %%sql magic command:

```
%load_ext sparksql_magic
```

Replace the following in each cell containing the %%sql magic command:

```
%%sql -> %%sparksql
```

Edit each path to the data make the change wasb:///data -> ../data:

~~flights = spark.read.csv('wasb:///data/raw-flight-data.csv', schema=flightSchema, header=True)~~
```
flights = spark.read.csv('../data/raw-flight-data.csv', schema=flightSchema, header=True)
```

~~airports = spark.read.csv('wasb:///data/airports.csv', header=True, inferSchema=True)~~
```
airports = spark.read.csv('../data/airports.csv', header=True, inferSchema=True)
```

etc.


Run Notebook

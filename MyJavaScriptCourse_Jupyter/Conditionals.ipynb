# Conditionals

```javascript
// if statement

function testNum1(num) {
    if (num > 0) {
        return "Positive";
    }
    return "Not positive";
}

console.log(testNum1(-5));
console.log(testNum1(0));
console.log(testNum1(5));
```

```javascript
// if-else statement

function testNum2(num) {
    if (num > 0) {
        result = "Positive";
    } else {
        result =  "Not positive";
    }
    return result;
}

console.log(testNum2(-5));
console.log(testNum2(0));
console.log(testNum2(5));
```

```javascript
// if-else-if-else statement

function testNum3(num) {
    if (num > 0) {
        result = "Positive";
    }
    else if (num< 0) {
        result = "Negative";
    }
    else {
        result = "Neutral";
    }
    return result;
}

console.log(testNum3(-5));
console.log(testNum3(0));
console.log(testNum3(5));
```

# Vectors and Matrices

## Vector Notation

### Row Vectors

$$\vec{v}=\begin{bmatrix}v_{1} v_{2}\end{bmatrix}$$

### Column Vectors

$$\vec{v}=\begin{bmatrix}v_{1}\\v_{2}\end{bmatrix}$$

### Vector Length

$$\|\vec{v}\|=\sqrt{v_{1}\;^{2}+v_{2}\;^{2}}$$

## Higher Dimentional Vectors

### Row Vectors

$$\vec{v}=\begin{bmatrix}v_{1} v_{2}…v_{n}\end{bmatrix}$$

### Column Vectors

$$\vec{v}=\begin{bmatrix}v_{1}\\v_{2}\\…\\v_{n}\end{bmatrix}$$

### Vector Length

$$\|\vec{v}\|=\sqrt{v_{1}\;^{2}+v_{2}\;^{2}+…+v_{n}\;^{2}}$$

## Vector Addition

$$\vec{v}+\vec{w}=\begin{bmatrix}v_{1}\\v_{2}\end{bmatrix}+\begin{bmatrix}w_{1}\\w_{2}\end{bmatrix}=\begin{bmatrix}v_{1}+w_{1}\\v_{2}+w_{2}\end{bmatrix}$$

## Vector Scalar Product

$$c \cdot \vec{v}=c \cdot \begin{bmatrix}v_{1}\\v_{2}\end{bmatrix}=\begin{bmatrix}c \cdot v_{1}\\c \cdot v_{2}\end{bmatrix}$$

## Vector Dot Product

$$\vec{v} \cdot \vec{w}=(v_{1} \cdot w_{1})+(v_{2} \cdot w_{2})+…+\;(v_{n} \cdot w_{n})$$

Also, where $\theta$ is the angle between $\vec{v}$ and $\vec{w}$

$$\vec{v} \cdot \vec{w}=\|\vec{v}\|\|\vec{w}\|\cos(\theta) $$ 

## Vector Transpose

$$
\begin{bmatrix}v_{1}v_{2}\end{bmatrix}^{T}
=
\begin{bmatrix}v_{1} \\ v_{2}\end{bmatrix}
$$

$$
\begin{bmatrix}v_{1} \\ v_{2}\end{bmatrix}^{T}
=
\begin{bmatrix}v_{1}v_{2}\end{bmatrix}
$$

## Matrix Notation

$$
A=
\begin{bmatrix}
  a_{1,1}&a_{1,2}&a_{1,3} \\
  a_{2,1}&a_{2,2}&a_{2,3} \\
\end{bmatrix}
$$

## Matrix Addition

$$
\begin{bmatrix}
a_{1,1} & a_{1,2} & a_{1,3} \\
a_{2,1} & a_{2,2} & a_{2,3} \\
\end{bmatrix} +
\begin{bmatrix}
b_{1,1} & b_{1,2} & b_{1,3} \\
b_{2,1} & b_{2,2} & b_{2,3} \\
\end{bmatrix} =
\begin{bmatrix}
a_{1,1}+b_{1,1} & a_{1,2}+b_{1,2} & a_{1,3}+b_{1,3} \\
a_{2,1}+b_{2,1} & a_{2,2}+b_{2,2} & a_{2,3}+b_{2,3} \\
\end{bmatrix}
$$

### Example: Matrix Addition

$$
\begin{bmatrix}
  1 & 2 & 3 \\
  4 & 5 & 6 \\
\end{bmatrix}
+
\begin{bmatrix}
  6 & 5 & 4 \\
  3 & 2 & 1 \\
\end{bmatrix}
=
\begin{bmatrix}
  7 & 7 & 7 \\
  7 & 7 & 7 \\
\end{bmatrix}
$$


## Matrix Scalar Product

$$
c \cdot 
\begin{bmatrix}
  a_{1,1} & a_{1,2} & a_{1,3} \\
  a_{2,1} & a_{2,2} & a_{2,3} \\
\end{bmatrix}
=
\begin{bmatrix}
  c \cdot a_{1,1} & c \cdot a_{1,2} & c \cdot a_{1,3} \\
  c \cdot a_{2,1} & c \cdot a_{2,2} & c \cdot a_{2,3} \\
\end{bmatrix}
$$

## Matrix Dot Product

$$
\begin{bmatrix}
  a_{1,1} & a_{1,2} & a_{1,3} \\ a_{2,1} & a_{2,2}&a_{2,3}
\end{bmatrix}
\cdot
\begin{bmatrix}
  b_{1,1} & b_{1,2} \\ b_{2,1} & b_{2,2} \\ b_{3,1} & b_{3,2}
\end{bmatrix}
=
\begin{bmatrix}
(a_{1,1}b_{1,1}+a_{1,2}b_{2,1}+a_{1,3}b_{3,1}) & (a_{1,1}b_{1,2}+a_{1,2}b_{2,2}+a_{1,3}b_{3,2}) \\(a_{2,1}b_{1,1}+a_{2,2}b_{2,1}+a_{2,3}b_{3,1}) & (a_{2,1}b_{1,2}+a_{2,2}b_{2,2}+a_{2,3}b_{3,2}) \\
\end{bmatrix}
$$

## Example: Matrix Dot Product

$$
\begin{bmatrix}1&2&3 \\ 4&5&6\end{bmatrix} \cdot
\begin{bmatrix}1&4 \\ 2&5 \\ 3&6\end{bmatrix}
=
\begin{bmatrix}(1+4+9)&(4+10+18)\\(4+10+18)&(16+25+36)\end{bmatrix}
=
\begin{bmatrix}14&32\\32&77\end{bmatrix}
$$

## Matrix Identity

$$\begin{bmatrix}1 & 0 & 0\\0 & 1 & 0\\0 & 0 & 1\end{bmatrix} \cdot A=A$$

### Example: Matrix Identity

$$\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}
\cdot
\begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}
=
\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$$

## Matrix Determinant

### 2x2 Matrix Determinant:

$$
\det
\begin{bmatrix}
  a_{1,1} & a_{1,2} \\
  a_{2,1} & a_{2,2} \\
\end{bmatrix}
=
a_{1,1}a_{2,2}-a_{1,2}a_{2,1}
$$

### 3x3 Matrix Determinant:

$$
\det\begin{bmatrix}
  a_{1,1} & a_{1,2} & a_{1,3}\\
  a_{2,1} & a_{2,2} & a_{2,3}\\
  a_{3,1} & a_{3,2} & a_{3,3}\\
\end{bmatrix}
=
{
a_{1,1}
\det
\begin{bmatrix}
  a_{2,2} & a_{2,3}\\
  a_{3,2} & a_{3,3}\\
\end{bmatrix}
}
-
{
a_{1,2}
\det
\begin{bmatrix}
  a_{2,1} & a_{2,3}\\
  a_{3,1} & a_{3,3}\\
\end{bmatrix}
}
+
{
a_{1,3}
\det
\begin{bmatrix}
  a_{2,1} & a_{2,2}\\
  a_{3,1} & a_{3,2}\\
\end{bmatrix}
}
$$

## Vector Cross Product in 3D

$$
\vec{v} \times \vec{w}
=
\begin{bmatrix}
  v_{1} \\
  v_{2} \\
  v_{3} \\
\end{bmatrix}
\times
\begin{bmatrix}
  w_{1} \\
  w_{2} \\
  w_{3} \\
\end{bmatrix}
= det
\begin{bmatrix}
  \hat{i} & \hat{j} & \hat{k}  \\
  v_{1} & v_{2} & v_{3}  \\
  w_{1} & w_{2} & w_{3}  \\
\end{bmatrix}
=
\hat{i}(v_{2}w_{3}-v_{3}w_{2})-\hat{j}(v_{1}w_{3}-v_{3}w_{1})+\hat{k}(v_{1}w_{2}-v_{2}w_{1})
$$

Also, where $\theta$ is the angle between $\vec{v}$ and $\vec{w}$, and $\vec{u}$ is the unit vector orthogonal to both $\vec{v}$ and $\vec{w}$:

$$\vec{v} \times \vec{w}=\|\vec{v}\|\|\vec{w}\|\sin(\theta)\vec{u}$$

## Matrix Transpose

$$
\begin{bmatrix}
  a & b \\
  c & d \\
\end{bmatrix}^{T}
=
\begin{bmatrix}
  a & c \\
  b & d \\
\end{bmatrix}
$$

$$
\begin{bmatrix}
  a & b & c \\
  d & e & f \\
\end{bmatrix}^{T}
=
\begin{bmatrix}
  a & d \\
  b & e \\
  c & f 
\end{bmatrix}
$$

$$
\begin{bmatrix}
  a_{1,1} & a_{1,2} & a_{1,3} \\
  a_{2,1} & a_{2,2} & a_{2,3} \\
  a_{3,1} & a_{3,2} & a_{3,3} \\
\end{bmatrix}^{T}
=
\begin{bmatrix}
  a_{1,1} & a_{2,1} & a_{3,1} \\
  a_{1,2} & a_{2,2} & a_{3,2} \\
  a_{1,3} & a_{2,3} & a_{3,3} \\
\end{bmatrix}
$$

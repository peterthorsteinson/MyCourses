# Equations

## Line plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xvals = np.arange(-10, +10, 0.01) # Grid of 0.01 spacing from -10 to 10
yvals = np.cos(xvals)  # Evaluate function on xvals
plt.plot(xvals, yvals) # Create line plot with yvals against xvals
plt.show() # Show the figure
```

```python
newyvals = 1 - 0.5 * xvals**2    # Evaluate quadratic approximation on xvals
plt.plot(xvals, newyvals, 'r--') # Create line plot with red dashed line
plt.title('Example plots')
plt.xlabel('Input')
plt.ylabel('Function values')
plt.show()
```

## Contour plots

```python
xlist = np.linspace(-2.0, 1.0, 100) # Create 1-D arrays for x,y dimensions
ylist = np.linspace(-1.0, 2.0, 100)
X,Y = np.meshgrid(xlist, ylist)     # Create 2-D grid xlist,ylist values
Z = np.sqrt(X**2 + Y**2)            # Compute function values on the grid
plt.contour(X, Y, Z, [0.5, 1.0, 1.2, 1.5], colors = 'k', linestyles = 'solid')
plt.show()
```

```python
xlist = np.linspace(-2.0, 1.0, 100) # Create 1-D arrays for x,y dimensions
ylist = np.linspace(-1.0, 2.0, 100)
X,Y = np.meshgrid(xlist, ylist)     # Create 2-D grid xlist,ylist values
Z = np.sqrt(X**2 + Y**2)            # Compute function values on the grid
plt.contour(X, Y, Z, [0.5, 1.0, 1.2, 1.5], colors = 'k', linestyles = 'solid')
plt.axes().set_aspect('equal')      # Scale the plot size to get same aspect ratio
plt.axis([-1.0, 1.0, -0.5, 0.5])    # Set axis limits
plt.show()
```

```python
x = np.linspace(0,100,100) # 100 linearly spaced numbers rom 0 to 100
y = (9/5)*(x)+32           # conversion function
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot(x,y, 'r')
plt.title('Celsius to Fahrenheit')
plt.xlabel('Celsius')
plt.ylabel('Fahrenheit')
plt.show() 
```

```python
x = np.linspace(-5,5,100) # 100 linearly spaced numbers
y = x**2                  # the function, which is y = x^2 here
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center') # setting the axes at centre
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.plot(x,y, 'r') # plot the function
plt.show()         # show the plot
```

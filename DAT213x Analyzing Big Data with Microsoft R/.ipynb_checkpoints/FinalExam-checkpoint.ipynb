Consider the ChickWeight dataset, which is part of Base R datasets. Which of the following code using rxHistogram() will produce similar result to hist(subset(ChickWeight,Time==21)$weight,freq=TRUE,breaks=10)?

```r
ChickWeight <- datasets::ChickWeight
```

```r
hist(subset(ChickWeight,Time==21)$weight,freq=TRUE,breaks=10)
```

```r
rxHistogram(~weight, ChickWeight, rowSelection = (Time==21), histType = "Percent", numBreaks = 10)
```

```r
rxHistogram(~weight, ChickWeight, rowSelection = (Time==21), numBreaks = 10)
```

```r
 rxHistogram(~weight, ChickWeight, numBreaks = 10)
```

```r
rxHistogram(~weight, ChickWeight)
```

Working Locally using Docker

        The original course describes running the labs in an Azure hosted notebook.
        But the following describes how to do it locally in a Docker container.

Can run these labs in jupyter notebooks in Azure or you can run them locally using Docker.
- See: https://realpython.com/pyspark-intro

DOS Shell:
```
> docker run -t -p 8888:8888 jupyter/pyspark-notebook
```
        Displays 3 URLs -> view 3rd link in local browser

DOS Shell (new one):
```
> docker container list
        Copy container ID to clipboard
> docker exec -it container-id bash
        Paste in correct container ID in above command
```
